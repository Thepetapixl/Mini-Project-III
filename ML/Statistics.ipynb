```python
# Importing Libraries to use Dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn import linear_model
```

```python
# Reading the csv dataset to perform operations on it

df = pd.read_csv(r"C:\Users\Rupeshwar\Desktop\Car_Parking.csv")
```

```python
# Finding out how many rows and columns does dataset have 
print(df.shape)
```

```python
# Finding out the entries which have given a rating of 4.5 and above

df.loc[df['Rating'] >= 4.5]
```

```python
#Plotting a bar Graph to indicate the performance of the system

df.plot(x="Year", y="Amount", kind="bar")
```

```python
# Addition of the Revenue collected in these Years

y_2016 = df.loc[df['Year'] == 2016, 'Amount'].sum()

y_2017 = df.loc[df['Year'] == 2017, 'Amount'].sum()

y_2018 = df.loc[df['Year'] == 2018, 'Amount'].sum()

y_2019 = df.loc[df['Year'] == 2019, 'Amount'].sum()

y_2020 = df.loc[df['Year'] == 2020, 'Amount'].sum()
```

```python
# Performance in terms of Revenue generated by the Parking System over the years
years = [y_2016,y_2017,y_2018,y_2019,y_2020]

my_labels = '2016','2017','2018','2019','2020'
my_colors = ['Red','Blue','Black', 'Yellow','Violet']
plt.pie(years,labels=my_labels,autopct='%1.1f%%')
plt.title('Revenue')
plt.axis('equal')
plt.show()
```

```python
# Displaying enteries in each year

addition = df['Year'].value_counts()
print(addition)
```

```python
# Adding all records(car enteries) in each years

cars_16 = df.loc[df.Year == 2016, 'Year'].count()
cars_17 = df.loc[df.Year == 2017, 'Year'].count()
cars_18 = df.loc[df.Year == 2018, 'Year'].count()
cars_19 = df.loc[df.Year == 2019, 'Year'].count()
cars_20 = df.loc[df.Year == 2020, 'Year'].count()
```

```python
# Performance in terms of Cars parked in the Parking System over the years
cars_year = [cars_16, cars_17, cars_18, cars_19, cars_20]

my_labels = '2016','2017','2018','2019','2020'
my_colors = ['Red','Blue','Black', 'Yellow','Violet']
plt.pie(cars_year,labels=my_labels,autopct='%1.1f%%')
plt.title('Cars Parked')
plt.axis('equal')
plt.show()
```
